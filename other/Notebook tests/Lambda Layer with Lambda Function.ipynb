Lambda Layer with Lamda Function

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import tensorflow.keras.backend as K
import tensorflow.keras.layers as KL
import tensorflow.keras.models as KM
```

```python
rank_4_tensor = tf.zeros([2, 7, 7, 6])
print(rank_4_tensor)
print(K.int_shape(rank_4_tensor))
```

Erstelle einen neuen Tensor mit den Wert der jeweiligen Dimension

```python
print(tf.shape(input=rank_4_tensor)[0])
print(tf.shape(input=rank_4_tensor)[1])
print(tf.shape(input=rank_4_tensor)[2])
print(tf.shape(input=rank_4_tensor)[3])
```

```python
print("nur reshape:", tf.reshape(rank_4_tensor, [tf.shape(input=rank_4_tensor)[0], -1, 2]))
```

```python
func = lambda t: tf.reshape(t, [tf.shape(input=t)[0], -1, 2])
print("nur lambda func:", func(rank_4_tensor))
```

```python
rpn_class_logits = KL.Lambda(
    lambda t: tf.reshape(t, [tf.shape(input=t)[0], -1, 2]))(rank_4_tensor)
print("lambda func + lambda layer:", rpn_class_logits)
```
